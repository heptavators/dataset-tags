## Menambahkan random tag pada talent yang jumlah tag < 5

```python
def impute_random_tag(tags: list) -> str:
    if len(tags) < 5:
        while len(tags) != 5:
            random_tag = np.random.choice(list_tags)
            tags.append(random_tag)

    for i in range(len(tags)):
        if tags[i] not in list_tags:
            tags[i] = np.random.choice(list_tags)

    for i in range(len(tags)):
        tag = tags[i].split("-")
        if len(tag) > 1:
            tags[i] = " ".join(tag)
            
    tags = np.array(tags)
    while len(tags) > 5:
        idx = np.random.randint(len(tags))
        tags = np.delete(tags, idx)
    return "|".join(tags)
```

```python
gform_cleaned.loc[:, 'hobi-cleaned'] = gform_cleaned.loc[:, "hobi-cleaned"].str.split("|").map(impute_random_tag)
```

```python
outlier = []
for user_tags in (
    gform_cleaned["hobi-cleaned"].str.split('|').values
):
    for tag in user_tags:
        if tag not in list_tags:
            outlier.append(tag)

np.array(list(set(outlier)))
```

```python
location_df.head()
```

```python
list_lokasi = location_df['Lokasi'].values
list_kota = location_df['Kota/Kabupaten'].values
list_provinsi = location_df['Provinsi'].values
```

```python
outlier = []
for user_location in gform_cleaned["user_location"].values:
    # for tag in user_tags:
    if user_location not in list_lokasi:
        outlier.append(user_location)

np.array(list(set(outlier)))
```

```python
np.random.choice(location_df['Lokasi'])
```

```python
def impute_location(location):
    if len(location) <= 1:
        return np.random.choice(list_lokasi)

    provinsi = location[0]
    kota = location[1]
    
    if provinsi not in list_provinsi:
        provinsi = np.random.choice(list_provinsi)
    
    if kota not in list_kota:
        kota = np.random.choice(location_df.loc[location_df['Provinsi'] == provinsi, 'Kota/Kabupaten'].values)
    
    return f'{provinsi}, {kota}'
```

```python
gform_cleaned.loc[:, 'user_location'] = gform_cleaned['user_location'].str.split(',').map(impute_location)
```

```python
outlier = []
for user_location in gform_cleaned["user_location"].values:
    if user_location not in list_lokasi:
        outlier.append(user_location)

np.array(list(set(outlier)))
```

### Merge google form data and generated data

```python
gender_choice = ['L', 'P']
```

```python
name_df = pd.read_csv('../data/e_msmhs_1.csv', sep=';')
```

```python
boys = name_df[name_df['GENDER'] == 'L']
girls = name_df[name_df['GENDER'] == 'P']
```

```python
boys_name = []
for fullname in boys['FULLNAME'].str.split().to_numpy():
    for n in fullname:
        boys_name.append(n)
```

```python
girls_name = []
for fullname in girls["FULLNAME"].str.split().to_numpy():
    for n in fullname:
        girls_name.append(n)
```

```python
def generate_random_identity(minimum_age: int = 16, maximum_age: int = 40):
    girl = np.random.randint(2)

    if not girl:
        gender = gender_choice[girl]
        fullname = []
        for _ in range(np.random.randint(2, 5)):
            name = np.random.choice(boys_name)
            while name in fullname:
                name = np.random.choice(boys_name)
            fullname.append(name.capitalize())
        birth_date = fake.date_of_birth(
            minimum_age=minimum_age, maximum_age=maximum_age
        )
        age = 2023 - birth_date.year

        return (
            " ".join(fullname),
            gender,
            f"{birth_date.day}/{birth_date.month}/{birth_date.year}",
            age,
        )
    else:
        gender = gender_choice[girl]
        fullname = []
        for _ in range(np.random.randint(2, 5)):
            name = np.random.choice(girls_name)
            while name in fullname:
                name = np.random.choice(girls_name)
            fullname.append(name.capitalize())
        birth_date = fake.date_of_birth(
            minimum_age=minimum_age, maximum_age=maximum_age
        )
        age = 2023 - birth_date.year

        return (
            " ".join(fullname),
            gender,
            f"{birth_date.day}/{birth_date.month}/{birth_date.year}",
            age,
        )
```

## Impute missing values

```python
columns = ["user_name", "user_gender", "user_birth_date", "user_age", "user_location"]
for i in range(len(merged_users)):
    id = uuid4()
    name, gender, birth_date, age = generate_random_identity()
    location = np.random.choice(list_lokasi)
    values = {
        "user_name": name,
        "user_gender": gender,
        "user_birth_date": birth_date,
        "user_age": float(age),
        "user_location": location,
    }

    merged_users.loc[i, "user_id"] = id

    for col in columns:
        if pd.isna(merged_users.loc[i, col]):
            merged_users.loc[i, col] = values[col]
```
